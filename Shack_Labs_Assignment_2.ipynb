# *Problem Statement:*

Using ML/DL techniques, match similar products from the Flipkart dataset with the Amazon dataset. Once
similar products are matched, display the retail price from FK and AMZ side by side. Please explore as
many techniques as possible before choosing the final technique.


Data set Link:
##### Amazon: https://www.dropbox.com/sh/aypq6h3254207bs/AACzMLvo-XtK9sYAAma6FW0la?dl=0&file_subpath=%2Famz_com-ecommerce_sample.csv&preview=DS+-+Assignment+Part+2+data+set.zip

##### Flipkart: https://www.dropbox.com/sh/aypq6h3254207bs/AACzMLvo-XtK9sYAAma6FW0la?dl=0&file_subpath=%2Fflipkart_com-ecommerce_sample.csv&preview=DS+-+Assignment+Part+2+data+set.zip

```python
import pandas as pd
```

```python
df1 = pd.read_csv("amz_com-ecommerce_sample.csv")
```

```python
df2 = pd.read_csv("flipkart_com-ecommerce_sample.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.columns
```

```python
cols_not_used = ['uniq_id','crawl_timestamp', 'product_url','product_category_tree','product_rating','overall_rating','product_specifications', 'is_FK_Advantage_product','image','brand','product_specifications']
```

```python
df1 = df1.drop(cols_not_used,axis='columns')
```

```python
df1.shape
```

```python
df1.head(10)
```

```python
df2 = df2.drop(cols_not_used,axis='columns')
```

```python
df2.head(10)
```

```python
df1.rename(columns= {'product_name' : "Product name in Amazon"}, inplace=True)
df1.rename(columns= {'retail_price' : "Retail Price in Amazon"}, inplace=True)
df1.rename(columns= {'discounted_price' : "Discounted Price in Amazon"}, inplace=True)
```

```python
df2.rename(columns= {'product_name' : "Product name in Flipkart"}, inplace=True)
df2.rename(columns= {'retail_price' : "Retail Price in Flipkart"}, inplace=True)
df2.rename(columns= {'discounted_price' : "Discounted Price in Flipkart"}, inplace=True)
```

```python
df = pd.concat([df1,df2],axis=1)
df.head()
```

```python
new_df = df.drop(['description', 'pid'], axis='columns')
```

```python
new_df.describe()
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(inplace=True)
```

```python
new_df = new_df[new_df['Product name in Amazon'] == new_df['Product name in Flipkart']]
```

```python
new_df.shape
```

```python
new_df.describe()
```

**new_df contains the final data of the matched product name from the both Amazon and Flipkart datasets**
